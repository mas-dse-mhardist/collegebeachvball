```python
import pandas as pd
import urllib
import re
```

```python
def find_avg_award(x):
    y = str(x)
    y = re.findall(r'\$(.*) \(', y)
    if len(y) == 0:
        y = re.findall(r'Est Award Amount \$(.*)\|', str(x))
    return y[0] if len(y) else None

def find_num_awards(x):
    y = str(x)
    y = re.findall(r' \((.*) award available', y)
    if len(y) == 0:
        y = re.findall(r'\|(.*) award', str(x))
    return y[0] if len(y) else None

def min_gpa(x):
    y = str(x)
    y = re.findall(r'GPA (.*)', y)

    return y[0] if len(y) else None

def academic_standing(x):
    y = str(x)
    y = re.findall(r'\-(.*) standing', y)
    if len(y) == 0:
        y = re.findall(r'\-(.*)student', str(x))
    if len(y) == 0:
        y = re.findall(r'\-(.*)undergraduate', str(x))
    return x if len(y) else None

def major(x):
    y = str(x)
    y = re.findall(r'ajor in(.*)', y)
    return x if len(y) else None

def resident(x):
    y = str(x)
    y = re.findall(r'\-(.*) resident', y)
    return x if len(y) else None

def need(x):
    y = str(x)
    y = re.findall(r'\-(.*) financial need', y)
    return x if len(y) else None

def interest(x):
    y = str(x)
    y = re.findall(r'\-(.*)interest', y)
    return x if len(y) else None

def leadership(x):
    y = str(x)
    y = re.findall(r'\-(.*)eadership', y)
    return x if len(y) else None

def community(x):
    y = str(x)
    y = re.findall(r'\-(.*)community', y)
    return x if len(y) else None


df_scholarships = pd.DataFrame.from_csv('pauahi.csv').reset_index()
df_scholarships = df_scholarships[df_scholarships['Description'] != 'Apply Now']

df_scholarships['Award Amount'] = df_scholarships['Description'].apply(lambda x: find_avg_award(x))
df_scholarships['Number Awards Available'] = df_scholarships['Description'].apply(lambda x: find_num_awards(x))
df_scholarships['Minimum GPA'] = df_scholarships['Description'].apply(lambda x: min_gpa(x))
df_scholarships['Academic Standing'] = df_scholarships['Description'].apply(lambda x: academic_standing(x))
df_scholarships['Major'] = df_scholarships['Description'].apply(lambda x: major(x))
df_scholarships['Hawaii Resident'] = df_scholarships['Description'].apply(lambda x: resident(x))
df_scholarships['Financial Need'] = df_scholarships['Description'].apply(lambda x: need(x))
df_scholarships['Interest'] = df_scholarships['Description'].apply(lambda x: interest(x))
df_scholarships['Leadership'] = df_scholarships['Description'].apply(lambda x: leadership(x))
df_scholarships['Community'] = df_scholarships['Description'].apply(lambda x: community(x))
df_scholarships
```
